# Text preprocessing 2 Continuation (Convert text to vectors)

#### Bag of Words (BoW) 
A simple way to convert text to vectors by counting the frequency of each word in a document. It creates a vector where each element represents the count of a specific word in the document.

#### TF-IDF (Term Frequency-Inverse Document Frequency)
TF-IDF is a more advanced technique that not only counts the frequency of words but also considers how important a word is in a document relative to a collection of documents. It gives higher weight to words that are unique to a document and lower weight to common words.

Fomula:

    Term Frequency TF = (Number of times a word appears in a document) / (Total number of words in the document)

    IDF = log(Total number of documents / Number of documents containing the word)

    TF-IDF = TF * IDF


Eg:

Doc 1: He is a good boy
Doc 2: She is a good girl
Doc 3: Boy and girl are good

After stop words:

Doc 1: good boy
Doc 2: good girl
Doc 3: boy girl good

#### Term Frequency -> Computed for sentence

                Doc 1      Doc 2      Doc 3
        good    1/2         1/2         1/3

        boy     1/2         0            1/3

        girl    0           1/2         1/3


#### Inverse document Frequency -> Computed for Words

        Words               IDF

        good               log(3/3)=0

        boy                log(3/2)

        girl               log(3/2)


#### TF - IDF

                f1              f2                     f3
                good            boy                     girl

        Doc 1       0               (1/2)*log(3/2)          0

        Doc 2       0               0                       (1/2)log(3/2)

        Doc 3       0               (1/3)log(3/2)           (1/3)log(3/2)
