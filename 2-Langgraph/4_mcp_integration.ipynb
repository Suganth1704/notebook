### MCP (Model context protocol)





What is MCP?  
Model Context Protocol helps LLMs interact with multiple tools to gather context and provide an appropriate
output for a given input.

* It lets you connect to tools like ArXiv, Wikipedia, RAG databases, etc.
* It enables interaction with other agents.

Why MCP?  
From my understanding:

* If you want to work with ArXiv, you could write custom integration code.
* The problem is that interacting with many tools means writing a lot of code, and if a particular tool
  changes you must update your integrations.
* MCP makes the integration simple and, in the event of a new version or change in a tool, MCP handles
  it without you rewriting everything.



#### Important components:

1) MCP host
2) MCP Client
3) MCP servers 
        |
        v
4) Tool provider (Arxiv)



#### Communication between the components:



1) Input will be given, MCP host will connect to MCP sever using MCP clinet  and get the tools that are integrated with the MCP servers and give it back to client.
2) MCP client will gives the question + tools to the LLM and LLM will make the decision, whcih tool to use.
3) LLM will call the tool and get the context and gives the answer to the client and client will give the answer to the user.
